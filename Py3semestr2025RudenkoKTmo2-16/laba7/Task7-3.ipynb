```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 1Ô∏è‚É£ –°–æ–∑–¥–∞–Ω–∏–µ DataFrame (–∏–Ω–¥–∏–≤–∏–¥—É–∞–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è)
data = {
    "–ò–º—è": ["–ê–Ω–Ω–∞", "–ë–æ—Ä–∏—Å", "–í–∏–∫–∞", "–ì—Ä–∏–≥–æ—Ä–∏–π", "–î–º–∏—Ç—Ä–∏–π"],
    "–í–æ–∑—Ä–∞—Å—Ç": [23, 30, 22, 25, 28],
    "–û—Ü–µ–Ω–∫–∞": [88, 92, 79, 85, 90],
    "–ß–∞—Å—ã_—É—á—ë–±—ã": [5, 7, 3, 4, 6]
}

df = pd.DataFrame(data)
```

```python
# 2Ô∏è‚É£ –ù–∞–∑–≤–∞–Ω–∏—è –∫–æ–ª–æ–Ω–æ–∫
print("–ù–∞–∑–≤–∞–Ω–∏–µ –∫–æ–ª–æ–Ω–æ–∫:", df.columns.tolist())
```

```python
# 3Ô∏è‚É£ –ò–Ω–¥–µ–∫—Å–∞—Ü–∏—è –ø–æ —Å—Ç–æ–ª–±—Ü—É
print("\n–ö–æ–ª–æ–Ω–∫–∞ '–í–æ–∑—Ä–∞—Å—Ç':")
print(df["–í–æ–∑—Ä–∞—Å—Ç"])
```

```python
# 4Ô∏è‚É£ –ú–µ—Ç–æ–¥ loc (–≤—ã–±–æ—Ä —Å—Ç—Ä–æ–∫–∏ –ø–æ –∏–Ω–¥–µ–∫—Å—É)
print("\n–°—Ç—Ä–æ–∫–∞ —Å –∏–Ω–¥–µ–∫—Å–æ–º 2:")
print(df.loc[2])
```

```python
# 5Ô∏è‚É£ –î–æ–±–∞–≤–ª–µ–Ω–∏–µ –Ω–æ–≤–æ–≥–æ —Å—Ç–æ–ª–±—Ü–∞
df["–ë–∞–ª–ª—ã_–∑–∞_–ø—Ä–æ–µ–∫—Ç—ã"] = [15, 20, 10, 12, 18]
print("\nDataFrame —Å –Ω–æ–≤—ã–º —Å—Ç–æ–ª–±—Ü–æ–º:")
print(df)
```

```python
# 6Ô∏è‚É£ –°—Ä–∞–≤–Ω–µ–Ω–∏–µ —Å—Ç–æ–ª–±—Ü–æ–≤ –∏ –ø–æ–ª—É—á–µ–Ω–∏–µ –±—É–ª–µ–≤–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è
print("\n–°—Ä–∞–≤–Ω–µ–Ω–∏–µ '–û—Ü–µ–Ω–∫–∞' > '–ë–∞–ª–ª—ã_–∑–∞_–ø—Ä–æ–µ–∫—Ç—ã':")
print(df["–û—Ü–µ–Ω–∫–∞"] > df["–ë–∞–ª–ª—ã_–∑–∞_–ø—Ä–æ–µ–∫—Ç—ã"])
```

```python
# 7Ô∏è‚É£ –û–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ –ø–æ –≤–µ—Ä—Ç–∏–∫–∞–ª–∏ (concat —Å ignore_index)
df2 = pd.DataFrame({
    "–ò–º—è": ["–ï–ª–µ–Ω–∞", "–ò–≥–æ—Ä—å"],
    "–í–æ–∑—Ä–∞—Å—Ç": [24, 29],
    "–û—Ü–µ–Ω–∫–∞": [91, 87],
    "–ß–∞—Å—ã_—É—á—ë–±—ã": [6, 5],
    "–ë–∞–ª–ª—ã_–∑–∞_–ø—Ä–æ–µ–∫—Ç—ã": [14, 16]
})
df_vertical = pd.concat([df, df2], ignore_index=True)
print("\n–í–µ—Ä—Ç–∏–∫–∞–ª—å–Ω–æ–µ –æ–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ:")
df_vertical
```

```python
# 8Ô∏è‚É£ –û–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ –ø–æ –≥–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª–∏ (concat –ø–æ axis=1)
df_horizontal = pd.concat([df, df2], axis=1)
print("\n–ì–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª—å–Ω–æ–µ –æ–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ:")
print(df_horizontal)
```

```python
# 9Ô∏è‚É£ –ú–µ—Ç–æ–¥ describe
print("\n–ú–µ—Ç–æ–¥ describe():")
print(df.describe())
```

```python
# üîü –ö—É–º—É–ª—è—Ç–∏–≤–Ω–∞—è —Å—É–º–º–∞ –ø–æ —Å—Ç–æ–ª–±—Ü—É '–û—Ü–µ–Ω–∫–∞'
df["–ö—É–º—É–ª—è—Ç–∏–≤–Ω–∞—è_—Å—É–º–º–∞_–æ—Ü–µ–Ω–æ–∫"] = df["–û—Ü–µ–Ω–∫–∞"].cumsum()
print("\nDataFrame —Å –∫—É–º—É–ª—è—Ç–∏–≤–Ω–æ–π —Å—É–º–º–æ–π –æ—Ü–µ–Ω–æ–∫:")
print(df)
```

```python
# 11 –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞
plt.figure(figsize=(8,5))
plt.plot(df["–ò–º—è"], df["–û—Ü–µ–Ω–∫–∞"], marker='o', label="–û—Ü–µ–Ω–∫–∞")
plt.plot(df["–ò–º—è"], df["–ë–∞–ª–ª—ã_–∑–∞_–ø—Ä–æ–µ–∫—Ç—ã"], marker='s', label="–ë–∞–ª–ª—ã –∑–∞ –ø—Ä–æ–µ–∫—Ç—ã")
plt.title("–ì—Ä–∞—Ñ–∏–∫ –æ—Ü–µ–Ω–æ–∫ –∏ –±–∞–ª–ª–æ–≤ –∑–∞ –ø—Ä–æ–µ–∫—Ç—ã")
plt.xlabel("–°—Ç—É–¥–µ–Ω—Ç")
plt.ylabel("–ë–∞–ª–ª—ã")
plt.legend()
plt.grid(True)
plt.show()
```
