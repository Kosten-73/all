```python
import matplotlib.pyplot as plt
import numpy as np

def plot_break_even(variable_cost, fixed_cost, price, label):
    Q = np.linspace(0, 2*planned_volume, 100)

    revenue = price * Q
    total_cost = fixed_cost + variable_cost * Q

    break_even_units = fixed_cost / (price - variable_cost)
    print(f"{label}: Точка безубыточности = {break_even_units:.2f} единиц")

    plt.plot(Q, revenue, label=f"Доход ({label})", linestyle='--')
    plt.plot(Q, total_cost, label=f"Затраты ({label})", linestyle='-')
    plt.scatter(break_even_units, price*break_even_units, color='red', zorder=5)
    plt.text(break_even_units, price*break_even_units + 10, f"{break_even_units:.0f}", color='red')


variable_cost1 = 20    # Переменные затраты на единицу
fixed_cost1 = 5000     # Постоянные затраты
price1 = 50            # Цена продажи
planned_volume = 300   # Планируемый объем

variable_cost2 = 25
fixed_cost2 = 4000
price2 = 60
planned_volume2 = 300

plt.figure(figsize=(10,6))

plot_break_even(variable_cost1, fixed_cost1, price1, label="Набор 1")
plot_break_even(variable_cost2, fixed_cost2, price2, label="Набор 2")

plt.title("Точка безубыточности компании (сравнение двух наборов данных)")
plt.xlabel("Количество продукции (шт.)")
plt.ylabel("Денежные средства ($)")
plt.grid(True)
plt.legend()
plt.show()
```
