```python
import numpy as np
from IPython.display import display, Markdown
```

```python
# Функция для красивого вывода
def md(text):
    display(Markdown(text))
```

```python
md("## Вариант 2 — Работа с матрицей NumPy (размерность 5×7)")
```

```python
# 1️⃣ Создаем матрицу 5x7 с индивидуальными значениями
matrix = np.array([
    [1, 3, 5, 7, 9, 11, 13],
    [2, 4, 6, 8, 10, 12, 14],
    [5, 10, 15, 20, 25, 30, 35],
    [3, 6, 9, 12, 15, 18, 21],
    [8, 16, 24, 32, 40, 48, 56]
])
```

```python
md("**1. Исходная матрица (5×7):**")
matrix
```

```python
# 2️⃣ Определяем размерность матрицы
shape = matrix.shape
md(f"**2. Размерность матрицы:** {shape[0]} строк × {shape[1]} столбцов")
```

```python
# 3️⃣ Выделяем размерность объектов (строки и столбцы)
rows, cols = matrix.shape
md(f"**3. Количество строк:** {rows}, **Количество столбцов:** {cols}")
```

```python
# 4️⃣ Индексация по колонкам (например, возьмём 2-й и 5-й столбцы)
indexed_columns = matrix[:, [1, 4]]
md("**4. Индексация по колонкам (2-й и 5-й):**")
indexed_columns
```

```python
# 5️⃣ Поэлементная операция (вариант 2 — вычитание, например вычтем 2)
subtracted = matrix - 2
md("**5. Поэлементное вычитание 2 из каждого элемента:**")
subtracted
```

```python
# 6️⃣ Применяем метод transpose (транспонирование)
transposed = matrix.transpose()
md("**6. Транспонированная матрица:**")
transposed
```
