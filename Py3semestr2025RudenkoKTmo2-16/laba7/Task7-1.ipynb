```python
import numpy as np
from IPython.display import display, Markdown
```

```python
# Вывод с форматированием
def md(text):
    display(Markdown(text))
```

```python
md("## Вариант 2 — Работа с вектором NumPy `[4, 6, 3, 23, 56, 1, 9, 8]`")
```

```python
# 1 создание вектора
vector = np.array([4, 6, 3, 23, 56, 1, 9, 8])
md("**1. Исходный вектор:**")
vector
```

```python
# 2 Индексация по элементам (3-й, 5-й, 7-й)
indexed_elements = vector[[2, 4, 6]]
md("**2. Индексация по элементам (индексы [2, 4, 6]):**")
indexed_elements
```

```python
# 3 Индексация по диапазону (с 2 по 5 включительно)
range_elements = vector[2:6]
md("**3. Индексация по диапазону (элементы 2–5):**")
range_elements
```

```python
# 4 Переворачиваем массив
reversed_vector = vector[::-1]
md("**4. Перевернутый вектор:**")
reversed_vector
```

```python
# 5 Операция присвоения (меняем элемент с индексом 3 на 100)
vector[3] = 100
md("**5. После присвоения (элемент 3 = 100):**")
vector
```

```python
# 6 Математическая операция (вариант 2 — деление)
divided = vector / 2
md("**6. Деление каждого элемента на 2:**")
divided
```

```python
# 7 Argmax и Argmin
argmax_idx = vector.argmax()
argmin_idx = vector.argmin()
md(f"**7. Argmax:** индекс {argmax_idx}, значение {vector[argmax_idx]}")
md(f"**7. Argmin:** индекс {argmin_idx}, значение {vector[argmin_idx]}")
```

```python
# 8 Метод append — добавляем 77
new_vector = np.append(vector, 77)
md("**8. После добавления элемента 77:**")
new_vector
```
